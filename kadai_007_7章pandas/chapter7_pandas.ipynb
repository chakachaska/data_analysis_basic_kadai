```python
import pandas as pd
import matplotlib.pyplot as plt
import japanize_matplotlib  # 日本語フォント対応

# サンプルデータを作成
df = pd.DataFrame({
    '商品番号': ['YY4HAAZR', 'Z4WO0IY', '1QJFQ8QY', 'MESUDWQ', '5GRE8W6X'],
    'カテゴリー': ['弁当', '飲料水', '弁当', '菓子類', '図書・新聞']
})

# カテゴリー列の出現頻度をカウント
category_counts = df['カテゴリー'].value_counts()

# カウント結果を棒グラフとして表示
category_counts.plot(kind='bar')
plt.title("カテゴリーの出現頻度")
plt.xlabel("カテゴリー")
plt.ylabel("出現頻度")
plt.show()

# 仮の注文数データを追加
df['注文数'] = [24, 15, 30, 5, 12]  # 仮のデータを設定

# 商品番号ごとの注文数の基本統計量を算出
order_stats = df.groupby('商品番号')['注文数'].describe()
print("\n商品番号ごとの注文数の統計量:")
print(order_stats)
```
